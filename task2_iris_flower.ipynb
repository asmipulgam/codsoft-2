```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris=pd.read_csv('/content/IRIS.csv')
iris.head()
```

```python
#renaming columns
iris=iris.rename(columns={'sepal_length':'sl','petal_length':'pl','petal_width':'pw'})
```

```python
#separating input output columns
data=iris.values

# slicing the matrices
X=data[:,0:4]
Y=data[:,4]
```

```python
print(X.shape)
print(X)
```

```python
print(Y.shape)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,Y, test_size=0.2)
print(X_train.shape)
print(X_train)
```

```python
print(y_test.shape)
print(y_test)
```

```python
print(X_test.shape)
print(X_test)
```

```python
print(y_train.shape)
print(y_train)
```

```python
from sklearn.svm import SVC

model_svc=SVC()
model_svc.fit(X_train,y_train)
```

```python
prediction1 = model_svc.predict(X_test)

#calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction1))
```

```python
# New data for prediction
X_new = np.array([[3, 2, 1, 0.2], [4.9, 2.2, 3.8, 1.1], [5.3, 2.5, 4.6, 1.9]])

# Predicting the sizes of the iris flowers
predicted_sizes = model_svc.predict(X_new)

# Output the predicted sizes
print(predicted_sizes)
```
